```python
#TO_EXCEL() METHOD
import pandas as pd
i = pd.DataFrame({
'name': ['john', 'smith', 'andy', 'smith'],
'age': [18, 22, 30, 25],
'marks': [89, 92, 61, 77],
'grade': ['a', 'a', 'c', 'b']})
file1 = pd.ExcelWriter('dataofuser.xlsx')
i.to_excel(file1)
file1.save()
print('Excel File Generated.')
```
